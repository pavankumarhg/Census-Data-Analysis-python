```python
# 2. Literacy Rate Analysis
# Calculating Literacy Rate for each district
census['Literacy_Rate'] = census['Literate'] / census['Population'] * 100

# Plotting Literacy Rate by State
plt.figure(figsize=(10, 6))
sns.barplot(x='State_name', y='Literacy_Rate', data=census, ci=None)
plt.xticks(rotation=90)
plt.title('Literacy Rate by State')
plt.ylabel('Literacy Rate (%)')
plt.show()
```

```python
# 3. Worker Distribution Analysis
# Calculating worker participation rate
census['Worker_Participation_Rate'] = census['Workers'] / census['Population'] * 100

# Plotting Worker Participation Rate by State with unique colors
plt.figure(figsize=(10, 6))
sns.barplot(x='State_name', y='Worker_Participation_Rate', data=census, 
            ci=None, palette='husl')  # Using 'husl' color palette for distinct colors
plt.xticks(rotation=90)
plt.title('Worker Participation Rate by State')
plt.ylabel('Worker Participation Rate (%)')
plt.show()
```

```python
# 4. Gender Distribution of Workers - Top 5 States by Male and Female Worker Participation Rates
# Calculating Male and Female Worker Participation Rates
census['Male_Workers_Rate'] = census['Male_Workers'] / census['Male'] * 100
census['Female_Workers_Rate'] = census['Female_Workers'] / census['Female'] * 100

# Get the top 5 states based on both Male and Female Worker Participation Rates
top_5_male = census.nlargest(5, 'Male_Workers_Rate')
top_5_female = census.nlargest(5, 'Female_Workers_Rate')

# Combine both top 5 male and female data
top_5_combined = pd.concat([top_5_male, top_5_female]).drop_duplicates().sort_values('Male_Workers_Rate', ascending=False)

# Plotting the combined top 5 states for both Male and Female Workers
plt.figure(figsize=(12, 6))

# Creating the barplot with hue to differentiate male and female rates
sns.barplot(x='State_name', y='Male_Workers_Rate', data=top_5_combined, label='Male Worker Participation Rate', color='blue')
sns.barplot(x='State_name', y='Female_Workers_Rate', data=top_5_combined, label='Female Worker Participation Rate', color='red')

# Adding the legend
plt.legend(loc='upper right')

# Final plot adjustments
plt.xticks(rotation=90)
plt.title('Top 5 States by Male and Female Worker Participation Rates')
plt.ylabel('Worker Participation Rate (%)')

plt.tight_layout()
plt.show()
```

```python
# 5. Education Level Analysis with Color and Sorted Data
# Sorting the data by Graduate Education Population in descending order
census_sorted = census.sort_values('Graduate_Education', ascending=False)

# Plotting the total graduate education population by state with colors (Sorted)
plt.figure(figsize=(10, 6))
sns.barplot(x='State_name', y='Graduate_Education', data=census_sorted, estimator=sum, ci=None, palette='coolwarm')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.title('Graduate Education Population by State (Sorted)')
plt.ylabel('Total Graduate Education Population')

# Show the plot
plt.tight_layout()  # To ensure proper spacing and layout
plt.show()
```
